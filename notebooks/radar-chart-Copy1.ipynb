### Different Y-Axis Chart

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('input/train.csv')
data.head()
```

```python
fig, ax1 = plt.subplots(dpi=100)

color = 'tab:red'
ax1.plot(data['PassengerId'], data['Age'], color=color)
ax1.set_xlabel('PassengerId')
ax1.set_ylabel('Age', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # Share the same x-axis

color = 'tab:blue'
ax2.plot(data['PassengerId'], np.log(data['Age'].cumsum()), color=color)
ax2.set_ylabel('Age', color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax1.grid()
fig.tight_layout()
plt.show()
```
